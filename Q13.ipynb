```python
import matplotlib.pyplot as plt
```

```python
mentions = [500, 505]
years = [2017, 2018]
```

```python
plt.bar(years, mentions,)
plt.xlabel('Year')
plt.ylabel('Mentions')
plt.title('Mentions per year')
plt.show()
```

```python
plt.bar(years, mentions)
plt.xlabel('Year')
plt.ylabel('Number of mentions')
plt.title('Mentions per year')
plt.ylim(0)
plt.show()
```
