```python
import matplotlib.pyplot as plt
```

```python
salaries_and_tenures = [(83000, 8.7), (88000, 8.1), (48000, 0.7),
(76000, 6), (69000, 6.5), (76000, 7.5), (60000, 2.5), (83000, 10),
(48000, 1.9), (63000, 4.2)]
less_than_two=[]
btw_two_and_five =[]
more_than_five = []
for sal,ten in salaries_and_tenures:
    if(ten<2):
        less_than_two.append(sal)
    elif(ten>=2 and ten <=5):
        btw_two_and_five.append(sal)
    else:
        more_than_five.append(sal)
avg_ltt = sum(less_than_two)/len(less_than_two)
avg_btf = sum(btw_two_and_five)/len(btw_two_and_five)
avg_mtf = sum(more_than_five)/len(more_than_five)
if(avg_ltt>avg_mtf):
    print("(The avrage salary for < 2) > (than average salary for > 5)")
else:
    print("(The avrage salary for < 2) < (than average salary for > 5)")
if(avg_ltt>avg_btf):
    print("(The average salary for < 2) > (The average sallary for b/w 2and 5)")
else:
    print("(The average salary for < 2) < (The average sallary for b/w 2and 5)")
if(avg_btf>avg_mtf):
    print("(The average sallary for b/w 2and 5) > (than average salary for > 5)")
else:
    print("(The average sallary for b/w 2and 5) < (than average salary for > 5)")
print("The average salary for < 2 = ",avg_ltt)
print("The average salary for > 5 = ",avg_mtf)
print("The average salary for between 2 and 5 = ",avg_btf)
```

```python
sal=[s[0] for s in salaries_and_tenures]
ten=[s[1] for s in salaries_and_tenures]
```

```python
plt.xlabel("experience(yrs)")
plt.ylabel("slary($)")
plt.title("Salary vs Years of Experience")
```
