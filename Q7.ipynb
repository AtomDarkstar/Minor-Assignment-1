```python
import matplotlib.pyplot as plt

text = input("Enter some text: ").lower()

freq = {}
for ch in text:
    if char.isalpha():
        if ch in freq:
            freq[ch] = freq.get(ch)+1
        else:
            freq[ch] = 1
```

```python
plt.bar(freq.keys(),freq.values());
plt.xlabel("Letters")
plt.ylabel("Frequency")
plt.title("Frequency of letters in the String");
plt.show();
```
