```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
max_temps = [17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18]
min_temps = [-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58]

plt.bar(months, max_temps,color= "red", label='Max Temperature')
plt.bar(months, min_temps,color= "blue", label='Min Temperature')

plt.xlabel('Months')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Extremes in a Region of India')

plt.legend()
plt.show()
```
