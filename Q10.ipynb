```python
import matplotlib.pyplot as plt
```

```python
male_age = [53,51,71,31,33,39,52,27,54,30,64,26,21,54,52,20,59,32]
female_age = [53,65,68,21,75,46,24,63,61,24,49,41,39,40,25,54,42,32,48,23,23]
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,5))

ax[0].hist(male_age)
ax[0].set_title('Male Age')
ax[0].set_xlabel('Age')
ax[0].set_ylabel('Frequency')

ax[1].hist(female_age)
ax[1].set_title('Female Age')
ax[1].set_xlabel('Age')
ax[1].set_ylabel('Frequency')
plt.show();
```
