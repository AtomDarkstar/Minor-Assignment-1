```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10,200)
y1 = x*np.sin(x)
y2 = x**2*np.sin(x)
y3 = x**3*np.sin(x)
y4 = x**4*np.sin(x)
```

```python
plt.plot(x, y1, label='xsin(x)')
plt.plot(x, y2, label='x^2sin(x)')
plt.plot(x, y3, label='x^3sin(x)')
plt.plot(x, y4, label='x^4sin(x)')

plt.title('sine functions plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show();
```
